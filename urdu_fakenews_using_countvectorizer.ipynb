```python
import os
import pandas as pd
```

```python
#i have data in textual form in a folder , where each news is an individual
#.txt file, i have to convert them into a single .csv file, so below is the 
#process 
import os
news=[]
for file in os.listdir('Train/Fake'):
    if file.endswith('.txt'):
        filepath=os.path.join('Train/Fake',file)
        #now opening each file and reading them
        with open(filepath,'r',encoding='utf-8') as f:
            sentence=f.read()
            news.append(sentence)


#now converting the file into pandas data frame
import pandas as pd
df1=pd.DataFrame(news,columns=['news'])
#above data is fake news so labeling them as 0 for fake
df1['label']=0
df1
```

```python
#now i am reading real news into new dataframe
news=[]
for file in os.listdir('Train/Real'):
    if file.endswith('.txt'):
        filepath=os.path.join('Train/Real',file)
        with open(filepath,'r',encoding='utf-8') as f:
            sentence=f.read()
            news.append(sentence)

df2=pd.DataFrame(news,columns=['news'])
df2['label']=1
df2
```

```python
#now concatenating two fake and real news data frame into single dataframe 
#along vertical because they have same columns
dfnew=pd.concat([df1,df2],axis=0)
dfnew
```

```python
#now saving this into .csv file
dfnew.to_csv("news.csv",index=False)
```

```python
dfnew.tail(5)
```

```python
#dropping if there is any empty row
dfnew=dfnew.dropna()
dfnew
```

```python
#now splitting x and y values 
x=dfnew['news'].values
y=dfnew['label']
x.shape
```

```python
#stopwords is urdu
stopwords = [
    "اور", "ہے", "میں", "کی", "کو", "سے", "کا", "یہ", "ایک", "کہ", "گی", 
    "گے", "ہیں", "ہوں", "پر", "نے", "کر", "دی", "یا", "تھا", "تھی", "تھے", 
    "تو", "اگر", "بھی", "کیے", "ہو", "والا", "والی", "والے", "جس", "جن", 
    "کیوں", "یہاں", "وہاں", "جنہیں", "جنہيں", "کرنا", "کرنے", "کرکے", 
    "ہونا", "ہونے", "ہوں", "کیونکہ", "تک", "تاکہ", "کرتا", "کرتی", 
    "کرتے", "کررہا", "کررہی", "کررہے", "کرچکا", "کرچکی", "کرچکے", 
    "کرچکا", "کرچکی", "کرچکے", "کرچکیں", "کروں", "کرو", "کرے", "کریں", 
    "کرتے", "کرتی", "کرتا", "رہا", "رہی", "رہے", "چکا", "چکی", "چکے", 
    "چکیں", "سکتا", "سکتی", "سکتے", "سکیں", "چاہیے", "ہوگی", "ہوگا", 
    "ہوں", "ہو", "ہے", "ہیں", "تھا", "تھی", "تھے", "ہوتا", "ہوتی", 
    "ہوتے", "ہوتیں", "تک", "تاکہ", "دو", "دوں", "دینا", "دی", "دے", 
    "دوں", "دیتا", "دیتی", "دیتے", "دیتیں", "ہوا", "ہوئی", "ہوئے", 
    "ہوا", "ہو", "ہوں", "ہے", "ہیں", "تھا", "تھی", "تھے", "میں", "کے", 
    "سے", "نے", "کہ", "یہ", "کی", "کا", "کو", "جو", "یہاں", "وہاں", "کیسے"
]
```

```python
def cleaned(x):
    import re 
    corpus = []
    for i in range(len(x)):
        cleaned=re.sub('[0-9a-zA-Z~!@#$%^&*()_+{}":?><,./;=]', ' ', x[i])
        cleaned=cleaned.split()
        cleaned=[word for word in cleaned if word not in stopwords]
        cleaned=' '.join(cleaned)
        corpus.append(cleaned)
    return corpus
    
corpus=cleaned(x)
print(type(corpus),len(corpus))
```

```python

from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(corpus,y,test_size=0.20,random_state=21)
print(type(trainx),len(testx))
```

```python
#now converting the cleaned text into arrays
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(ngram_range=(1,3))
trainx=count.fit_transform(trainx).toarray()
     
```

```python
trainx.shape
```

```python
testx=count.transform(testx)
```

```python
testx.shape
```

```python
#now applying mutlinomial naive bayes algorithm
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
```

```python
trainy.shape
```

```python
try:
    nb.fit(trainx,trainy)
    print("model has been successfully trained")
except Exception as e:
    print("training unsuccessfull")
```

```python
predictedy=nb.predict(testx)
print(predictedy)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("accuracy of our model is {:.2f}%".format(accuracy_score(testy,predictedy)*100))
print("\nclassification report is \n",classification_report(testy,predictedy))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.heatmap(confusion_matrix(testy,predictedy),annot=True,fmt='d',cmap='Greens')
plt.title("confusion matrix of model")
plt.xlabel("predicted label")
plt.ylabel("true label")
plt.show()
```

```python
trainy.shape
```

```python
#passive aggressive is another algorithm for text classification, it is
#robust to noisy data because they can adjust their model parameters incrementally
from sklearn.linear_model import PassiveAggressiveClassifier
pac=PassiveAggressiveClassifier()
pac.fit(trainx,trainy)
```

```python
predictedy=pac.predict(testx)
predictedy
```

```python
print("accuracy of our model is {:.2f}%".format(accuracy_score(testy,predictedy)*100))
print("\nclassification report is \n",classification_report(testy,predictedy))
plt.figure(figsize=(10,5))
sns.heatmap(confusion_matrix(testy,predictedy),annot=True,fmt='d',cmap='Greens')
plt.title("confusion matrix of model")
plt.xlabel("predicted label")
plt.ylabel("true label")
plt.show()
```
